### Data Preprocessing

```python
feature_data.select_dtypes(include=object).columns
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder

label_encoder=LabelEncoder()
preprocess_data=feature_data.copy()
```

```python
preprocess_data["Education_Label"]=label_encoder.fit_transform(feature_data["Education"])
```

```python
preprocess_data.head()
```

```python
preprocess_data["Marital_Status"].values.shape
```

```python
marital_values=preprocess_data["Marital_Status"].values.reshape(-1,1)
one_encoder=OneHotEncoder(handle_unknown="ignore",sparse=False)
```

```python
encoder_marital=one_encoder.fit_transform(marital_values)
encoder_marital
```

```python
one_encoder.get_feature_names_out(["Marital_Status"])
```

```python
marital_encoder_df=pd.DataFrame(encoder_marital,columns=one_encoder.get_feature_names_out(["Marital_Status"]))
marital_encoder_df.head()
```

```python
preprocess_data=pd.concat([preprocess_data,marital_encoder_df],axis=1)
preprocess_data.head()
```

```python
preprocess_data=preprocess_data.select_dtypes(exclude=object)
preprocess_data.head()
```

```python
preprocess_data=preprocess_data.drop(columns=["Recency","Complain","Response"])
preprocess_data.head()
```

```python
preprocess_data.dropna(inplace=True)
```

```python
preprocess_data=preprocess_data.reset_index(drop=True)
```

```python
preprocess_data.isnull().sum()
```

```python
preprocess_data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

for value in ["Spent","Income","Age"]:
  preprocess_data[value]=scaler.fit_transform(preprocess_data[value].values.reshape(-1,1)).round(2)

preprocess_data.describe()
```

```python
preprocess_data.drop(columns="Dt_Customer",inplace=True)
```

```python
preprocess_data.info()
```

### DIMENSIONALITY REDUCTION



```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
components_data=pca.fit_transform(preprocess_data)
reduction_data=pd.DataFrame(components_data,columns=["PC1","PC2","PC3"])
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection="3d")
x,y,z=reduction_data["PC1"],reduction_data["PC2"],reduction_data["PC3"]
ax.scatter(x,y,z,c="green",marker="o")
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
elbow=KElbowVisualizer(KMeans(),k=10)
elbow.fit(reduction_data)
elbow.show()
```

```python
kmeans=KMeans(n_clusters=5)
kmeans.fit(reduction_data)
clusters=kmeans.fit_predict(reduction_data)
```

```python
fig=plt.figure(figsize=(10,8))
x,y,z=reduction_data["PC1"],reduction_data["PC2"],reduction_data["PC3"]
sns.scatterplot(x=x,y=y,hue=clusters,data=reduction_data,palette="viridis")
```

```python
preprocess_data["Clusters"]=clusters
```

```python
id2labels={1:"very_low_spend",2:"low_spend",3:"moderate_spend",0:"high_spend",4:"very_high_spend"}
```

```python
preprocess_data["labels"]=preprocess_data["Clusters"].map(id2labels)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
import pickle 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(
    preprocess_data.drop(columns=["labels","Clusters"]),
    preprocess_data["Clusters"],
    random_state=42,
    test_size=0.2
)

kneighbors=KNeighborsClassifier(algorithm="brute",n_neighbors=5,p=2,weights="distance")

kneighbors.fit(x_train,y_train)
labels_pred=kneighbors.predict(x_test)

print(f1_score(y_test,labels_pred,average="weighted"))

```
